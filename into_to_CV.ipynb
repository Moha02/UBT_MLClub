```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

https://www.docs.opencv.org/master/

```python
image = cv2.imread(r'pic.jpeg')
```

```python
image.shape
```

```python
image = cv2.resize(image, (800, 600))
image.shape
```

```python
type(image)
```

```python
plt.imshow(image)
```

```python
fiximg = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(fiximg)

```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.imshow(fiximg)
```

```python
face_detector = cv2.CascadeClassifier('haarcascade_face.xml')
```

```python
detections = face_detector.detectMultiScale(fiximg)
```

```python
detections
```

```python
len(detections)
```

```python
for (x, y, w, h) in detections:
  #print(x, y, w, h)
  cv2.rectangle(fiximg, (x, y), (x + w, y + h), (0,255,255), 5)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.imshow(fiximg)
```

```python
saudi = cv2.imread(r'saudi.jpg')
saudifiximg = cv2.cvtColor(saudi,cv2.COLOR_BGR2RGB)
plt.imshow(saudifiximg)
```

```python
detections = face_detector.detectMultiScale(saudifiximg)
```

```python
detections
```

```python
len(detections)
```

```python
for (x, y, w, h) in detections:
  #print(x, y, w, h)
  cv2.rectangle(saudifiximg, (x, y), (x + w, y + h), (0,255,255), 5)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.imshow(saudifiximg)
```
